```python
#@title Load model
%%capture
from transformers import AutoTokenizer, AutoModelForSequenceClassification, BertForSequenceClassification

model_name = "Silicon23/BERTForDetectingDepression-Twitter2015" # @param ["Silicon23/BERTForDetectingDepression-Twitter2015", "Silicon23/BERTForDetectingDepression-Twitter2020", "Silicon23/BERTForDetectingDepression-Twitter2015-baseBERT"] {allow-input: true}

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
```

```python
#@title Inference
import torch
import math
text = "" #@param{type:"string"}
inputs = tokenizer(text, truncation=True, return_tensors="pt", padding="max_length")

model.eval()
with torch.no_grad():
    logits = model(**inputs).logits

s0, s1 = logits[0].cpu().numpy()
s0, s1 = (math.exp(s0), math.exp(s1))
# print(s1/(s0+s1))


predicted_class_id = logits.argmax().item()
id2label = {0: "Not Depressed", 1: "Depressed"}
print(id2label[predicted_class_id])
```
